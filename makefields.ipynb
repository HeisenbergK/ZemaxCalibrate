```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
npoints = 48
```

```python
def pol2cart(phi):
    x = np.cos(phi)
    y = np.sin(phi)
    return x, y
```

```python
phis = np.linspace(0,2*np.pi-(2*np.pi/npoints),npoints)
qs = []
us = []
for i in phis:
    a, b = pol2cart(i)
    qs.append(a)
    us.append(b)
qu = []

for i in range(len(qs)):
    qu.append("%.4f, %.4f\n"%(qs[i],us[i]))

x = []
y = []
xy = []
thetas = []
for i in range(len(qs)):
    theta = 0.5*np.arctan2(us[i],qs[i])
    if (theta<0.0):
        theta+=np.pi
    xfield = np.cos(theta)
    yfield = np.sin(theta)
    x.append(xfield)
    y.append(yfield)
    thetas.append(theta)
    xy.append("%.4f, %.4f\n"%(xfield,yfield))
```

```python
plt.clf()
plt.plot(qs,us,'.')
plt.xlabel("q")
plt.ylabel("u")
plt.show()
```

```python
plt.clf()
plt.plot(x,y,'.')
plt.xlabel("Ex")
plt.ylabel("Ey")
plt.show()
```

```python
plt.clf()
plt.plot(np.multiply(thetas,(180.0/np.pi)),'.')
plt.ylabel(r"$\theta$")
plt.show()
```
